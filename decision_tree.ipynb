```python
import pandas as pd
```

```python
data=pd.read_csv("heart.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
x=data.drop("target",axis=1)
y=data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
# for saving the model 
import joblib 
joblib.dump(model, 'dtModel.pkl')
```

```python
from sklearn.tree import plot_tree
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=[35,25])
plot_tree(model)
plt.show()
```

```python
# post pruning
# tells best parameters of model so that we get the best result
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
params={
    'criterion':['gini', 'entropy'],
    'max_depth':[5,6,7,8,9,10]
}
```

```python
model= DecisionTreeClassifier()
```

```python
gridCV= GridSearchCV(model,param_grid=params,cv=5)
```

```python
gridCV.fit(x_train,y_train)
```

```python
gridCV.best_params_
```

```python
# model with new params..
pre_model= DecisionTreeClassifier(criterion='entropy',max_depth=10) 
pre_model.fit(x_train,y_train)
y_pred= pre_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=[35,25])
plot_tree(pre_model)
plt.show()# shows numbers vgaira
```
