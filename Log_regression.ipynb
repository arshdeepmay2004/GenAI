```python
# Logistic regression is a fundamental supervised machine learning algorithm used for classification,
#  predicting the probability of a binary outcome (e.g., yes/no, 0/1) based on independent variables
```

```python
import pandas as pd
data=pd.read_csv("student_data.csv")
data
```

```python
data.drop("student_id", axis=1,inplace=True)
```

```python

data.drop("family_income", axis=1,inplace=True)
```

```python
data.grade_category.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
# used in place of "map" to manupulate data like yes:1 no:0
encoder=LabelEncoder()
```

```python
encoder
```

```python
data.grade_category=encoder.fit_transform(data.grade_category)
```

```python
data.pass_fail=encoder.fit_transform(data.pass_fail)
```

```python
data.tutoring=encoder.fit_transform(data.tutoring)
```

```python
data.study_environment=encoder.fit_transform(data.study_environment)
```

```python
data.parental_education=encoder.fit_transform(data.parental_education)
```

```python
data.internet_access=encoder.fit_transform(data.internet_access)
```

```python
data.gender=encoder.fit_transform(data.gender)
```

```python
data.head()
```

```python
x=data.drop("grade_category",axis=1)
y=data.grade_category
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score ,mean_squared_error
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_pred,y_test)
```
